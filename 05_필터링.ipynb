```python
# isin() 매서드 활용
# isin() 매서드를 활용한 필터링:DataFrame객체.isin(추출값의 리스트)

import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")

pd.set_option("display.max_columns", 10) # 10개만 보임 / 실제적인데이터는 15개임 하단 참조

mask1 = titanic["embark_town"] == "Southampton"
mask2 = titanic["embark_town"] == "Queenstown"

df_boolean = titanic[mask1 | mask2]
df_boolean.head()
```

```python
mask3 = titanic["embark_town"].isin(["Southampton","Queenstown"]) #조건식

titanic[mask3]
```
