```python
import pandas as pd
```

```python
dict_data = {
    "a" : 1,
    "b" : 2,
    "c" : 3,
}

sr = pd.Series(dict_data)

print(type(sr))
print(sr)
```

```python
# 인덱스 배열
## Series객체.index

# 데이터값 배열
## Series객체.values

print(sr.index)
print(sr.values)
```

```python
# 시리즈 인덱스

import pandas as pd

list_data = ["2010-01-02", 3.14, "ABC", 100, True]

sr = pd.Series(list_data)

print(type(sr))
print(sr)
```

```python
print(sr.index)
print(sr.values)
print(sr.dtype)
```

```python
# 배열의 크기
print(len(sr))
```

```python
# shape
sr.shape
```

```python
# ndim
print(sr.ndim)
```

```python
import pandas as pd

tup_data = ("영인","2010-05-01","여","Ture")
sr = pd.Series(tup_data, index=["이름","생년월일","성별","학생여부"])
print(sr)
```

```python
# 시리즈 원소를 선택

print(sr["이름"])
```

```python
print(sr[["생년월일","성별"]])
```

```python
print(sr)
```

```python
print(sr["생년월일":"성별"])
```

```python
# 빈시리즈

sr = pd.Series()
print(sr)
```

```python
sr = pd.Series(5)
print(sr)
```

```python
sr = pd.Series(5,index=["a","b","c"])
print(sr)
```
