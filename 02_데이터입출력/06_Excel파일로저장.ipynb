```python
import pandas as pd

data1 = {'name' : [ 'Jerry', 'Riah', 'Paul'],
    'algol' : [ "A", "A+", "B"],
    'basic' : [ "C", "B", "B+"],
    'c++' : [ "B+", "C", "C+"],
}

data2 = {
    "c0":[1,2,3],
    "c1":[4,5,6],
    "c2":[7,8,9],
    "c3":[10,11,12],
    "c4":[13,14,15]
}

#df1 지정 인덱스를 네임기준으로
df1 = pd.DataFrame(data1)
df1 = df1.set_index("name") # 재할당
print(df1)
print("\n")

#df2 지정 , 인덱스를 c0 기준
df2 = pd.DataFrame(data2)
df2 = df2.set_index("c0") # 재할당
print(df2)

# with 써서 현재위치 폴더에 엑셀파일 쓰게끔 만들어서 시트1,시트2에 각각 df1,df2 넣을 이름 지정해주기

with pd.ExcelWriter("./data/df_excelwriter.xlsx") as writer:
    df1.to_excel(writer, sheet_name="sheet1")
    df2.to_excel(writer, sheet_name="sheet2")
```
