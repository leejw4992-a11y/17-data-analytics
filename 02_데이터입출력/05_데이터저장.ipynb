```python
# csv 파일로 저장

# DataFrame객체.to_csv("이름")
import pandas as pd


data = {'name' : [ 'Jerry', 'Riah', 'Paul'],
    'algol' : [ "A", "A+", "B"],
    'basic' : [ "C", "B", "B+"],
    'c++' : [ "B+", "C", "C+"],
}

df = pd.DataFrame(data)
df.set_index("name")
df
```

```python
df.to_csv("./data/df_sample.csv")
```

```python
# json 파일로 저장

import pandas as pd


data = {'name' : [ 'Jerry', 'Riah', 'Paul'],
    'algol' : [ "A", "A+", "B"],
    'basic' : [ "C", "B", "B+"],
    'c++' : [ "B+", "C", "C+"],
}

df = pd.DataFrame(data)
df = df.set_index("name")
df
```

```python
# DataFrame객체.to_json("파일이름")

df.to_json("./data/df_sample.json")
```

```python
df.to_json("./data/df_sample_split.json", orient="split")
```

```python
df.to_json("./data/df_sample_records.json",orient="records")
```

```python
df.to_json("./data/df_sample_index.json", orient="index")
```
