```python
# 행 인덱스 변경 : DataFrame객체.rename(index = {기존 인덱스:새 인덱스})
# 열 인덱스 변경 : DataFrame객체.rename(columns = {기존 이름:새이름})

import pandas as pd

matrix = [
    [15,"남","덕영중"],
    [17, "여","수리중"]
]

df = pd.DataFrame(matrix, index =["준서","예은"],columns=["나이","성별","학교"] )
df
```

```python
df = df.rename(columns = {
    "나이":"연령",
    "성별":"남녀",
    "학교":"소속"
    })
df

# 리네임 하고난 후에는 무조건 재할당해줘야함!
```

```python
# 행과 열의 위치 바꾸기

# 행, 열 바꾸기 : DataFrame객체.transpose() 또는 T

import pandas as pd

exam_data = {
    "이름" : ['서준','우현','민아'],
    "수학" : [90,80,70],
    "영어" : [98,89,95],
    "음악" : [85,95,100],
    "체육" : [100,90,90]
}

df = pd.DataFrame(exam_data)

df
```

```python
df = df.transpose()
df
```

```python
df = df.T
df
```
