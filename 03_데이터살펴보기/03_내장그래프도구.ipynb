```python
import pandas as pd
```

```python
# 선 그래프

# DataFrame객체.plot()

df = pd.read_excel("./data/남북한발전전력량 (2).xlsx")
df.head()
```

```python
df_ns = (df.iloc[[0,5], 3:]) #0번과 5번쨰 행가져오고, 3번열부터 끝까지 가져오라
df_ns.index=["South","North"]
df_ns
```

```python
df_ns.info()
```

```python
# 막대그래프

## 막대그래프 : DataFrame객체.plot(kind="bar")

df = pd.read_excel("./data/남북한발전전력량 (2).xlsx")

df_ns = df.iloc[[0,5],3:]
df_ns
```

```python
df_ns.index = ["South","North"]
df_ns
```

```python
tdf_ns = df_ns.T
tdf_ns.head(2)
```

```python
tdf_ns.plot(kind="bar")
```

```python
tdf_ns.plot(kind="barh")
```

```python
# 히스토그램

tdf_ns
```

```python
tdf_ns = tdf_ns.astype(int)
tdf_ns.info()
```

```python
tdf_ns.plot(kind="hist")
```

```python
tdf_ns.plot(
    kind="hist",
    alpha=0.5,
    figsize=(4,1),
    grid=True,
    color="skyblue",
    edgecolor="black"
            )
```

```python
tdf_ns.hist(figsize=(6,2)) #각각 north와 south 히스토그램을 분류해서 그림.
```

```python
tdf_ns.hist(bins=20, grid=False, figsize=(6,2))
```
