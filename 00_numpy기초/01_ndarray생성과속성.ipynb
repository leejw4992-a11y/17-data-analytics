```python
# 배열 속성 (답설명 :1차원 값이고 4개의 값을 가지고있다.)

arr1 = np.array([1,2,3,4])
arr1.shape
```

```python
# 해석 : 2행 3열

arr2 = np.array([
    [1,2,3],
    [4,5,6]
])
arr2.shape
```

```python
arr3 = np.array([
    [[1,2],
    [3,4]],
    [[5,6],
    [7,8]]
])

arr3.shape
```

```python
# ndim 
print(arr1.ndim) # 1차원 배열 (결과값 기준으로 해석)
print(arr2.ndim) # 2차원 배열 (결과값 기준으로 해석)
print(arr3.ndim) # 3차원 배열 (결과값 기준으로 해석)
```
